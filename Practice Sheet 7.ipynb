```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```
